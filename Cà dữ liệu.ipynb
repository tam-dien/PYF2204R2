```python
import requests as req

def laydulieu(ngay):
    web = req.get('https://www.thantai.net/xo-so-truyen-thong/'+str(ngay)+'-07-2022')
    chuoi = web.text
    
    bd = 0
    L_tag = []
    for i in range(len(chuoi)):
        if chuoi[i] == "<":
            bd = i
        if chuoi[i] == ">":
            if chuoi[bd:i+1] not in L_tag:
                L_tag.append(chuoi[bd:i+1])
    
    for i in L_tag:
        chuoi = chuoi.replace(i,"\n")
        
    chuoi = chuoi.split()
    kq = []
    for i in chuoi:
        if i.isnumeric() and len(i) == 5:
            kq.append(i)
            
    return kq[0]

chuoi = ""

for i in range(1,32):
    chuoi += laydulieu(i) + "\n"

f = open("xoso.txt","w")
f.write(chuoi)
f.close()

print(chuoi)
```
