# Vòng lặp For

### Vòng lặp for

Cú pháp:

`for <biến> in <tập lặp>:
    //Lệnh 1
//Lệnh 2`

- Dòng for sẽ lặp với số lần là số lượng phần tử trong tập lặp
- Mỗi lần lặp, <biến> sẽ lấy giá trị của mỗi phần từ trong tập
- Kiểu dữ liệu của biến phụ thuộc vào kiểu dữ liệu của phần tử trong tập lặp

### lệnh range

Cú pháp:

`range(a,b,c)`

Lệnh này giúp tạo ra một tập hợp có giá trị từ a cho đến b - 1 với bước nhảy là c

Ví dụ:

`range(0,5,1)`

Lệnh này sẽ giúp tạo ra tập {0,1,2,3,4}

`range(1,10,2)`

Lệnh này sẽ giúp tạo ra tập {1,3,5,7,9}

##### Lưu ý

#### Nếu ta chỉ điền một tham số, giá trị đó sẽ là giá trị b, a sẽ mặc định là 0 và c sẽ mặc định là 1

Ví dụ:
`range(3)`

Lệnh này sẽ giúp tạo ra tập {0,1,2}

#### c không được phép bằng 0, ta có thể bỏ trống c và chỉ điền 2 tham số a và b, khi ấy c sẽ mặc định là 1

Ví dụ:
`range(0,5)`

Lệnh này sẽ giúp tạo ra tập {0,1,2,3,4}

#### Các thông số a, b, c PHẢI là số nguyên và được phép âm

Ví dụ:
`range(0,-5,-1)`

Lệnh này sẽ giúp tạo ra tập {0,-1,-2,-3,-4}

#### Nếu giá trị của c dương, thì a phải lớn hơn b, ngược lại nếu giá trị của c âm, thì a phải bé hơn b. Nếu không, lệnh range sẽ trả về tập rỗng

Ví dụ:
`range(0,-5,2)`

Lệnh này sẽ giúp tạo ra tập {}

### Vòng lặp for với chuỗi

Tập lặp còn có thể là chuỗi

Khi ấy, biến sẽ truy xuất từng phần tử của chuỗi

Ví dụ:

Hãy nhập vào chuỗi, sau đó in ra từng phần tử của chuỗi

### Các lệnh tương tác với vòng lặp

Các lệnh tương tác với vòng lệnh while như `break`, `continue` có thể tương tác với vòng lặp for

Khác với while, dòng lặp for không thể bị mắc lỗi lặp vô hạn lần (do không có tập lặp nào có giá trị vô tận)
